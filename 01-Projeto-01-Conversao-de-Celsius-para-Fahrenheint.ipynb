# Projeto 01 - Conversao de Celsius para Fahrenheint
Usando Deep Learning iremos resolver o problema de conversão de Celsius para Fahrenheint

### Definição do problema
Equação para conversão = *T(°F) = T(°C) x 9/5 + 32

### Importando as bibliotecas


```python
import tensorflow as tf
tf.__version__
```

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn')
```

```python
#importação da base de dados
temperatura_df = pd.read_csv('datasets/Celsius-to-Fahrenheit.csv')
temperatura_df.reset_index(drop=True, inplace=True)
```

```python
temperatura_df.head()
```

```python
sns.scatterplot(x='Celsius', y='Fahrenheit', data=temperatura_df);
```

```python
X_train = temperatura_df['Celsius']
y_train = temperatura_df['Fahrenheit']
```

### construção e treinamento do modelo

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(units=1, input_shape=[1]))
model.summary()
```

```python
from tensorflow.keras.optimizers import Adam
model.compile(optimizer=Adam(0.01), loss='mean_squared_error')
```

```python
epoch_hist = model.fit(X_train, y_train, epochs=500)
```

### Avaliação do modelo

```python
epoch_hist.history.keys()
```

```python
plt.plot(epoch_hist.history['loss'], label='Training loss')
plt.title('Model Loss Progess During Train')
plt.xlabel('Epochs')
plt.ylabel('Training Loss')
plt.legend('Training loss')
```
