```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
```

```python
plt.style.use('seaborn')
```

```python
diabets = pd.read_csv('datasets/diabetes.csv')
diabets.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(diabets.drop('Outcome', axis=1),
                                                   diabets['Outcome'], random_state=21, test_size=.3)
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
mms = MinMaxScaler()
X_train_scaled = mms.fit_transform(X_train)
X_test_scaled = mms.transform(X_test)
```

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(units=100, activation='relu', input_shape=(8,)))
model.add(tf.keras.layers.Dense(units=250, activation='softmax'))
model.add(tf.keras.layers.Dense(units=500, activation='relu'))
model.add(tf.keras.layers.Dropout(0.5))
model.add(tf.keras.layers.Dense(units=500, activation='softmax'))
model.add(tf.keras.layers.Dense(units=1000, activation='relu'))
model.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

```

```python
model.compile(optimizer="Adam", loss='binary_crossentropy', metrics='accuracy')
```

```python
epochs_hist = model.fit(X_train_scaled, y_train, epochs=250)
```

```python
pred_tf = model.predict(X_test_scaled)
pred_tf = (pred_tf > 0.5)
```
